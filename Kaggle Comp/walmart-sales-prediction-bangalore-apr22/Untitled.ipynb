```python
df=pd.read_csv('train.csv')
```

```python
# Null values.
df.isnull().sum()/len(df)*100
```

```python
Womenme=df[df['ProductCategory']=='WomenClothing'].mean()['Sales(In ThousandDollars)']
Menme=df[df['ProductCategory']=='MenClothing'].mean()['Sales(In ThousandDollars)']
othernme=df[df['ProductCategory']=='OtherClothing'].mean()['Sales(In ThousandDollars)']
print(Womenme)
print(Menme)
print(othernme)
```

```python
pd.read_excel('WeatherData.xlsx')
```

```python
df[df['ProductCategory']=='WomenClothing']['Sales(In ThousandDollars)'].replace(np.nan,Womenme,inplace=True
df[df['ProductCategory']=='MenClothing']['Sales(In ThousandDollars)'].replace(np.nan,Menme,inplace=True)
df[df['ProductCategory']=='OtherClothing']['Sales(In ThousandDollars)'].replace(np.nan,othernme,inplace=True)
```

```python
df.fillna(method='ffill',inplace=True)
```

```python
df.isnull().sum()
```

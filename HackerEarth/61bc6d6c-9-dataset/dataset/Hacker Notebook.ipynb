```python
import pandas as pd
pd.set_option('display.max_columns',None)
```

```python
# Training set
train=pd.read_csv('train.csv')
train.head(5)
```

```python
# Testing Set
test=pd.read_csv('test.csv')
test.head(5)
```

```python
# Rename
test.rename(columns={'MACDHist':'MAC_Hist'},inplace=True)
```

```python
# Dropping of columns
train.drop(columns=['ID'],inplace=True)
test.drop(columns=['ID'],inplace=True)
```

```python
# Drop the featuure
train.drop(columns=['Date'],inplace=True)
test.drop(columns=['Date'],inplace=True)
```

```python
# split of Dataset.
x_train,y_train=train.drop(columns=['Price']),train['Price']
```

```python
#Test
x_test=test.copy()
```

```python
# Separation cat and num
cat=x_train.select_dtypes(exclude='number').columns
num=x_train.select_dtypes(include='number').columns
print(cat)
print(num)
```

```python
x_test.columns
```

```python
# check for null values
x_train.isnull().sum()
```

```python
# Encoding
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
```

```python
x_train['Company ']=oe.fit_transform(x_train[['Company ']])
x_test['Company ']=oe.transform(x_test[['Company ']])
```

```python
x_train.head()
```

```python
# Deal with the null values.
from sklearn.impute import KNNImputer
knn=KNNImputer()
```

```python
knn.fit(x_train,y_train)
```

```python
# Transform
tr=knn.transform(x_train)
te=knn.transform(x_test)
```

```python
X_train=pd.DataFrame(data=tr,columns=x_train.columns)
X_test=pd.DataFrame(data=te,columns=x_test.columns)
```

```python
X_train.isnull().sum().sum()
```

```python
# Transformer
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer()
```

```python
for i in num:
    X_train[i]=pt.fit_transform(X_train[[i]])
    X_test[i]=pt.transform(X_test[[i]])
```

```python
X_train.head()
```

```python
# Model Building

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
linreg=LinearRegression()
decision=DecisionTreeRegressor()
Random=RandomForestRegressor()
```

```python
# add Constant
```

```python
# Linear Model
linreg.fit(X_train,y_train)
```

```python
linreg.score(X_train,y_train)
```

```python
from sklearn import metrics
metrics.get_scorer_names()[:10]
```

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(estimator=Random,X = X_train,y=y_train,scoring='neg_root_mean_squared_error',cv=5)
score.mean()
```

```python
# Decision Tree
decision.fit(X_train,y_train)
```

```python
print('R-squared: ',decision.score(X_train,y_train))
```

```python
from sklearn.ensemble import AdaBoostRegressor,GradientBoostingRegressor
from sklearn.linear_model import Ridge,Lasso
```

```python
print('R-squared: ',Random.score(X_train,y_train))
```

```python
 # Pipeline
from sklearn.pipeline import Pipeline
pipe = Pipeline((
("lr", DecisionTreeRegressor(max_depth=10)),
))
```

```python
pipe.fit(X_train,y_train)
```

```python
print('R:Squared : ',pipe.score(X_train,y_train))
```

```python
pipe = Pipeline((
("pt",PowerTransformer()),
("lr", Lasso(alpha=1)),
))
```

```python
pipe.fit(X_train,y_train)
```

```python
print('R:Squared : ',pipe.score(X_train,y_train))
```

```python
from sklearn.preprocessing import PolynomialFeatures
pipe = Pipeline((
("pt", PowerTransformer()),
("poly",PolynomialFeatures(degree=2)),
("lr", LinearRegression())
))
```

```python
pipe.fit(X_train,y_train)

print('R:Squared : ',pipe.score(X_train,y_train))
```
